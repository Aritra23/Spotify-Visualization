```python
import csv
import matplotlib.pyplot as plt
import requests
import pandas as pd
import numpy as np
```

```python
file = 'spotify_data_lat_long.csv'
```

```python
spotify_df = pd.read_csv(file)
spotify_df
```

```python
spotify_update_df = pd.DataFrame(spotify_df.city_coord.str.split(', ',1).tolist(),
                                   columns = ['lat','long'])
spotify_update_df
```

```python
spotify_update_df['long'] = spotify_update_df['long'].map(lambda x: str(x)[:-1])
spotify_update_df
```

```python
spotify_update_df['lat'] = spotify_update_df['lat'].map(lambda x: str(x)[1:])
spotify_update_df
```

```python
spotify_update_df.to_csv('lat_long.csv')
```
